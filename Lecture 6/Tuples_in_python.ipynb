# Tuples
A tuple in Python is Similar to a list. The difference between the two is that we cannot Change the elements Of a tuple once it is assigned whereas we can change the
elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics:
-  Ordered
- Unchangeble
-  Allows duplicate

# Creating Tuples

```python
l = [] #list
t = () #tuple
```

```python
# empty tuple
t1 = ()
print(t1)
print(type(t1))
```

```python
t2 = ("Abdullah",) #comma is necessary for single element tuple
print(t2)
print(type(t2))
```

```python
# homo
t3 = (1, 2, 3, 4, 5)
print(t3)

# hetero
t4 = (1,2,3 "Hello", 3.4, True[1,2,3])
print(t4)

# nested tuple
t5 = (1,2, ("Hello", "World"), (3.4, 5.6))
print(t5)

#using type conversion function
t6 = tuple([1,2,3,4,5])
print(t6)
```
