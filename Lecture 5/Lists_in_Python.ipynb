## What are Lists?
List is a data type where you can store multiple items under 1 name. More technically, lists act like dynamic arrays which means you can add more items on the fly.


### Array vs Lists
 


```python
L = [1,2,3]
```

```python
print(L)
print(type(L))
print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
```

```python
print(id(1))
print(id(2))
print(id(3))
```

## Characterstics of a List
- Ordered
- Changeble/Mutable
- Hetrogeneous
- Can have duplicates
- are dynamic
- can be nested
- items can be accessed
- can contain any kind of objects in python

```python
L1 = [1,2,3]
L2 = [2,1,3]

L1 == L2 # will return False because the order is different
L1 is L2 # will return False because they are different objects in memory
```

```python
L1 = [1, 2, 3, 3]
print(L1)
```

```python
L1 = [1,2,3,3,4,5]
print(L1)
```

```python
#Empty List
print([])
#1D
print([1,2,3,4,5])
#2D
print([[1,2,3],[4,5,6],[7,8,9]])
#3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
#Heterogeneous List
print([1, "Hello", 3.4, True, [1,2,3]])
#Using type conversion function
print(list((1,2,3,4,5)))
type(list((1,2,3,4,5)))
```

# Accessing Items from a List
- indexing
- slicing

```python
l = [1, 2, 3, 4, 5]
```

```python
l[2]  # Accessing the 3rd element (index 2)
```

```python
l[-3] # Accessing the 3rd last element (index -3)
```

```python
l2 = [1,2,3,[4,5,6]]
```

```python
l2[3][1]    # Accessing the nested list [4,5,6]
```

```python
l2 [-1][-1]  # Accessing the last element of the nested list
```

```python
l3 = [[[1,2],[3,4]],[[5,6],[7,8]]]
```

```python
len(l3)  # Length of the outermost list
```

```python
l3[1][0][1]  # Accessing element '6' from the 3D list
```

```python
l3[-1][-1][-2]
```

```python
l = [1,2,3,4,5]
```

```python
l[0:2]  # Slicing to get first two elements
```

```python
l[:] # Slicing to get the entire list
```

```python
l[::2]  # Slicing to get every second element
```

```python
l[::-1] # Slicing to get the list in reverse order
```

```python
l3 = [[[1,2],[3,4]],[[5,6],[7,8]]]
```

```python
new_l3 = l3[1][0]
```

# Lists Methods/Functions
Lists is Mutable, That means you can change the list

```python
# append

L = [1,2,3,4]
print(L)
```

```python
L.append(5) # Appending 5 to the list
```

```python
print(L)
```

```python
L.append(500)  
```

```python
# editing with slicing

L =  [1,2,3,4,5]

L[1:4] = [20,30,40] # Replacing elements at index 1 to 3

print(L)
```

```python
L = [1,2,3,4,5]

L.append([6,7,8])
print(L)
```

# Extend


```python

L = [1,2,3,4,5]

a = [10,11,12]

L.extend(a)
print(L)
```

```python
L.extend("Hello")
print(L)
```

# Insert

```python
L = [1,2,3,4,5]

L.insert(1, 200)  # Inserting 200 at index 1

print(L)
```

```python
# Deleting elements from a list

L = [1,2,3,4,5,200]

del L[5]  # Deleting element at index 5

print(L)
```

```python
L = [1,2,3,4,5]
del L[1:3]  # Deleting elements from index 1 to 1
print(L)
```

# Remove

```python
L = [1,2,3,4,5]

L.remove(4)  # Removing element with value 4

print(L)
```

# Pop


```python
L = [1,2,3,4,5]
L.pop()  # Popping the last element
L.pop(2)  
print(L)
```

# Clear

```python
L = [1,2,3,4,5]
L.clear()
print(L)
```
