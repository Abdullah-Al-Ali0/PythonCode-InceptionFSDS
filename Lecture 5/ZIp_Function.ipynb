## Zip
The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each
passed iterator are paired together.




If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
L1 = [1,2,3,4,5]
L2 = [6,7,8,9,10]
```

```python
list(zip(L1,L2))
```

```python
L1 = [1,2,3,4,5]
L2 = [6,7,8,9]

list(zip(L1,L2))
```

```python
roll_no = [100, 101,102,103,104]
name = ['Abdullah', 'Alice', 'Bob', 'Charlie', 'David']

list(zip(roll_no, name))
```

```python
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i-j for i,j in zip(L1,L2)]
```

```python
for i,j in zip(L1,L2):
    print(i-j)

    
```

```python
L1 = [1,2,3,4,5]
L2 = [6,7,8,9,10]
L3 = [11,12]
list(zip(L1,L2,L3))
```
