# Exercise: print triangle coordinates

```python
row = int(input("Enter number of rows: "))

for i in range(1, row+1):
    for j in range (1, i+1):
        print("_", end= " ")
    
    print("")

```

```python
row = int(input("Enter number of rows: "))

for i in range(1, row+1):
    for j in range (1, i+1):
        print("*", end= " ")
    
    print("")

```

# Loop Control Statements
- Break
- Continue
- pass

```python
# break
for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
name = ["Alex", "Mamum", "Bappy" , "Abdullah", "Ridoy"]
for i in name:
    if i=="Abdullah":
        print(i)
        break
```

```python
# Continue

for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
# pass
for i in range(1,11):
    if i==5:
        pass
    print(i)
```
