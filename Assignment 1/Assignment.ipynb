# Assignment 2: ATM Withdrawal System

```python
balance = int(input("Please enter your account balance: ")) #Shouldve used float for more precision
widthdraw = int(input("Please enter the amount to withdraw: ")) #Shouldve used float for more precision
newBalance = balance - widthdraw

if widthdraw > balance:
    print("Insufficient Balance")

elif widthdraw <= balance :
    print("Transaction Successful.")
    print("Your remaining balance is : ", newBalance)
```

# Assignment 3: Student Grade Calculator


```python
total_marks = input("Please enter the total marks to check your grade(0-100): ")
total_marks = int(total_marks) # Also couldve used float for more precision

if total_marks >= 90 and total_marks <= 100:
    print("Congratulations, Your grade is A+.")

elif total_marks >= 80 and total_marks <= 89:
    print("Good Job, Your grade is A.")

elif total_marks >= 70 and total_marks <= 79:
    print("Graet, Your grade is B, Keep it up.")

elif total_marks >= 60 and total_marks <= 69:
    print("Not bad, Your grade is C, You can do better.")

elif total_marks < 60: # and total_marks==100:
    print("You failed, Your grade is F, try to improve next time.")

else:
    print("Possibly invalid Input, Please enter marks between 0-100.")
```

## Assignment 7: Traffic Light System


```python
color = input("What is the color of the traffic light(Options: red, yellow, or green).").lower() #converted to lowercase to avoid case sensitivity issues, it was a good idea

if color == "red":
    print("Stop")
elif color == "yellow":
    print ("Ready to go.")
elif color == "green":
    print("Go")
else:
    print("Invalid color.")
```

# Assignment 8: Leap Year Checker


```python
year = int(input("Please enter a year to check if it's a leap year or not: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print(" Not a Leap Year")
```
